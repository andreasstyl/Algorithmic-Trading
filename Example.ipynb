## The moving average crossover strategy

```python
s_lookback = 45 # defining the shorter lookback period
l_lookback = 90 # defining the longer lookback period
```

```python
df['s_lookback'] = df['Adj Close'].rolling(window = s_lookback).mean()
df['l_lookback'] = df['Adj Close'].rolling(window = l_lookback).mean()
```

```python
df[['Adj Close', 's_lookback','l_lookback']].plot(grid = True, linewidth = 1)
```

Everytime the 2 lookbacks corss we want to take a position

```python
df['position'] = (df['s_lookback'] > df['l_lookback'])
df['position']
```

```python
# better way so we can put what we want
df['position'] = np.where((df['s_lookback'] > df['l_lookback']),1, -1 )
df['position']
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
# now we plot evrything
df[['Adj Close','s_lookback','l_lookback','position']].plot(secondary_y = 
    'position', grid = True, linewidth = 1)
```

## Results of our strategy

```python
# what our strategy gave us
# buy and hold strategy
df['bnh_returns'] = np.log(df['Adj Close'] / df['Adj Close'].shift(1))

#strategy returns
df['strategy_returns'] = df['bnh_returns'] * df['position'].shift(1)

df[['bnh_returns','strategy_returns']].cumsum().plot(grid = True, linewidth = 1)
```
